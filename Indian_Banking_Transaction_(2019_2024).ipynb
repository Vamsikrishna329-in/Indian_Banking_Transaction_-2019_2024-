```python
df = pd.read_csv('indian_banking_transactions.csv')
print(df.head())
```

```python
df.info()
```

```python
print(df.isnull().sum())
df = df.fillna(0)
```

```python
transaction_amount = df[df['transaction_amount']>2500]
print(transaction_amount)
```

```python
transaction_amount = df[df['transaction_amount']>10000]
print(transaction_amount)
```

```python
transaction_type = df[df['transaction_type'] == 'Net_Banking']
print(transaction_type)
```

```python
average_net_banking_transaction = transaction_type['transaction_amount'].mean()
print(f"The average transaction amount for Net_Banking is: {average_net_banking_transaction:.2f}")
```

```python
average_transaction_by_type = df.groupby('transaction_type')['transaction_amount'].mean().reset_index()
print("Average transaction amount by transaction type:")
print(average_transaction_by_type.sort_values(by='transaction_amount', ascending=False))
```

```python
transaction_type = df[df['transaction_type'] == 'UPI']
print(transaction_type)
```

```python
average_UPI_transaction = transaction_type['transaction_amount'].mean()
print(f"The average transaction amount for UPI is: {average_UPI_transaction:.2f}")
```

```python
most_used_transaction_type = df['transaction_type'].value_counts().idxmax()
count_most_used = df['transaction_type'].value_counts().max()
print(f"The most used transaction type is '{most_used_transaction_type}' with {count_most_used} transactions.")
```

```python
least_used_transaction_type = df['transaction_type'].value_counts().idxmin()
count_least_used = df['transaction_type'].value_counts().min()
print(f"The least used transaction type is '{least_used_transaction_type}' with {count_least_used} transactions.")
```

```python
rtgs_transactions = df[df['transaction_type'] == 'RTGS']
max_rtgs_amount = rtgs_transactions['transaction_amount'].max()
min_rtgs_amount = rtgs_transactions['transaction_amount'].min()

print(f"Maximum transaction amount for RTGS: {max_rtgs_amount:.2f}")
print(f"Minimum transaction amount for RTGS: {min_rtgs_amount:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='transaction_type', y='transaction_amount', data=average_transaction_by_type.sort_values(by='transaction_amount', ascending=False), palette='viridis', hue='transaction_type', legend=False)
plt.title('Average Transaction Amount by Transaction Type')
plt.xlabel('Transaction Type')
plt.ylabel('Average Transaction Amount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
transaction_status_frequency = df['transaction_status'].value_counts()
print("Frequency of each transaction status:")
print(transaction_status_frequency)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(transaction_status_frequency, labels=transaction_status_frequency.index, autopct='%1.1f%%', startangle=140)
plt.title('Frequency of Transaction Statuses')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
transaction_status_by_account_type = pd.crosstab(df['account_type'], df['transaction_status'])
print("Relationship between Transaction Status and Account Type:")
display(transaction_status_by_account_type)
```

```python
total_transactions_per_account = transaction_status_by_account_type.sum(axis=1)
percentage_failed_by_account_type = (transaction_status_by_account_type['Failed'] / total_transactions_per_account) * 100

print("Percentage of 'Failed' transactions for each account type:")
display(percentage_failed_by_account_type.round(2))
```

```python
failed_transactions = df[df['transaction_status'] == 'Failed']
failed_transactions_by_channel = failed_transactions['channel'].value_counts()

most_failed_channel = failed_transactions_by_channel.idxmax()
num_failed_transactions = failed_transactions_by_channel.max()

print(f"The channel with the highest number of failed transactions is '{most_failed_channel}' with {num_failed_transactions} failed transactions.")
```
